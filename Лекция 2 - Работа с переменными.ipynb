# Переменные и работа с ними - лекция 2

Инициализация переменных - способ создания информации внутри программного кода. 


Переменные бывают типов: string, int, float, bool

Чтобы узнать тип переменной в коде достаточно вызвать функцию **type(value)**, гду value - переменная

```python
intVal = 9
stringVal = "Devyat"
floatVal = 9.0
boolVal = True
```

```python
print(
"1 - {0} - {1} \n".format(intVal, type(intVal))+
"2 - {0} - {1} \n".format(stringVal, type(stringVal)) +
"3 - {0} - {1} \n".format(floatVal, type(floatVal)) +
"4 - {0} - {1}".format(boolVal, type(boolVal))
)
```

Чаще всего пременные будут задаваться не в коде. С помощью **input("sample text")** можно приостановить программу для ожидания от пользователя ввода значения.

```python
newVal = input("Введите значение: \n")
print(newVal, type(newVal))
```

Мы ввели в прошлом примере 9 как число, однако Python определил переменную newVal как строку. Это произошло потому, что **input()** возвращает строковый тип данных. 
Мы можем манипулировать типами данных при помощи функций приведений: **int(value), float(value), bool(value), str(value)** 

```python
value = input("Сейчас все что вы введете - будет строкой. \n")
print("Значение - {0} Тип - {1} \n".format(value, type(value)))


value = int(value)
print("Значение - {0} Тип - {1} \n".format(value, type(value)))

value = float(value)
print("Значение - {0} Тип - {1} \n".format(value, type(value)))

value = str(value)
print("Значение - {0} Тип - {1} \n".format(value, type(value)))

value = bool(value)
print("Значение - {0} Тип - {1} \n".format(value, type(value)))
```

Приведение типов необходимо для работы с данными. Например мы должны обработать числовые данные, которые получаем из строки. Перед их обработкой мы приведем их в тип числовой. Например решим простую задачу. 

**Получим от пользователя количество однотипных товаров их стоимость и скидку пользователя в этом магазине. Вычислить полную стоимость покупки.**

```python
price = float(input()) #приведем к float, так как цена может быть не целым числом
count = int(input()) #приведем к int, так как количество всегда целое число
discount = float(input()) #приведем к float, так как скидки 
# первая часть под получение данных и предобработку. 
```

```python
summ = (1-discount/100)*price*count
# вторая часть кода - обработка данных
print("Сумма {0} предметов по {1} каждый - {2}".format(count,price, summ))
# последняя часть кода - отображение 
```

С числовым типом int и float можно совершать математические манипуляции: +,-,*,/,//,% 

Если подключать дополнительные библиотеки, например math, станут доступны sqr, sqrt и подобные функции

Строковый переменные - являются "коллекциями" символьного типа. Означает, что строки поддерживают индексацию.

```python
stringVal = "this is new string"
print(
    stringVal[0], "\n"+
    stringVal[0:7], "\n"+
    stringVal[0:10:2], "\n"
)
```

Подсчет массивов от 0 упрощает вычисление адреса памяти каждого элемента.

Если массив хранится в данной позиции в памяти (это называется адресом), позиция каждого элемента может быть вычислена как 

**element(n) = address + n * size_of_the_element**

Если вы считаете первый элемент первым, вычисление становится

**element(n) = address + (n-1) * size_of_the_element**
