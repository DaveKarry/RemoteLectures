# Базовые структуры алгоритмов - циклы

Предположим, у нас есть приложение, которое, выдает какой-то секрет, если ввести правильный пароль. 

```python
password = "password"
secret = "loops will help solve the problem"

checkPassword = input("Введите пароль: ")
if checkPassword == password:
    print(secret)
else:
    print("incorrect password")
```

Если мы введем значение пароля с ошибкой или не правильный пароль, то программа завершится, не дав нам попробовать ввести его повторно. 

Если мы захотим все-таки ввести второй пароль, нам необходимо будет запустить программу снова.

Пользователь обычно ленив и не хочет совершать лишние действия, особенно запускать приложения по-новому или вводить по-новой данные. Однако программист может использовать циклы, для того, чтобы дать возможность пользователю повторить ввод данных.

![loop1](images/4lec/loop1.png)

Как видим, пользователь не остается внутри этого алгоритма надолго. Однако мы можем поставить **условный цикл while**

```python
password = "password"
secret = "loops will help solve the problem"

checkPassword = input("Введите пароль: ")

while checkPassword != password:
    print("Вы ввели неверный пароль!")
    checkPassword = input("Введите пароль: ")

print("Вы ввели верный пароль!")
print(secret)
```

Теперь наша блок-схема выглядит так: 
    
![loop2](images/4lec/loop2.png)

Как видно из примера, код "возвращается" в блок сравнения, где он "начинает новый цикл". Поэтому цикл **While** называется условным циклом, ведь он проверяет выражение на **true/false**, точно так-же как и условный оператор **if-else**

Классические задачи с таким типом циклов:
    
* ограничение заполнения элементов в массиве(списке)

* повторение каких-либо действий до необходимого результата

Следующий тип циклов - итерационный, технически наш прошлый цикл с счетчиком был интерационно-условным, но все эти разделения являются необязательными и стоит только запоминать принцип работы. 

Предположим у нас есть список, из которого нужно вытащить значения в новый список. Например, вытащим значения больше чем 10 в новый список, в третий список удвоенные значения из первого списка, а в четвертый - только четные.

```python
oldList = [9,4,5,12,0,4, 5,4,89, 32]
bigList = []
doubleList = []
oddList = []
```

```python
for element in oldList:
    if element >10:
        bigList.append(element)
print(bigList)
```

```python
for element in oldList:
    doubleList.append(element*2)
print(doubleList)
```

```python
for element in oldList:
    if element %2 ==0:
        oddList.append(element*2)
print(oddList)
```

Если переводить дословно, получится нечто вроде: 

**ДЛЯ КАЖДОЙ** переменной **ВНУТРИ** списка:

    Делай штуки

Этот цикл обязательно пройдется по всем элементам внутри списка, поэтому н называется итерационным, ведь основной пункт это итерация, тоесть шаг внутри цикла.

```python
for i in range(4):  # равносильно инструкции for i in 0, 1, 2, 3:
    # здесь можно выполнять циклические действия
    print(i)
```

**range()** - Функция, которая создает отрезок с заданными параметрами.
Например **range(N)** -  создаст отрезок из N элементов с 0 до N-1

**range(a, b)** - составляет отрезок от a до b-1 

**range(a,b,c)** - составляет отрезок от a до b не включая шаг, c шагом в размер C.

```python
for i in range(2,7): 
    print(i)
```

```python
for i in range(-2,10, 2): 
    print(i)
```

# Задачи 

1) Дан список чисел, нужно за один цикл разделить этот список на список четных чисел и нечетных чисел.  Список задается случайными числами.

```python
from random import randint

numbers = []
for i in range(20):
    numbers.append(randint(-10, 10))
    
print(numbers)
```

```python
evenList = []
oddList = []
```

```python
for element in numbers:
    if element%2==0:
        evenList.append(element)
    else:
        oddList.append(element)
        
print(oddList)
print(evenList)
```
