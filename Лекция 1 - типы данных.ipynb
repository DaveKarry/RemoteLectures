## Типы данных в питоне

### Числовой тип
числа в питоне означают количественное значение, первый простой тип - int

```python
a = 10
b = 5
type(a),type(b)
```

С числами можно выполнять различные математические операции.



```python
summ = a+b
multiply = a*b
diff = a-b
quotient = a/b
degree = a**b
```

```python
summ,multiply,diff,quotient, degree
```

Не всегда результатом целочисленных выражений будет целое число.

```python
type(summ),type(multiply),type(diff),type(quotient), type(degree)
```

Тип float - тип данных для вещественных чисел.

```python
a = 0.3
b = 0.5
c = 5
```

действия над float и int проводятся одинаковые: 

```python
summ = a+b+c
multiply = a*b*c
diff = a-b-c
quotient = a/b/c
degree = a**b**c
```

```python
summ,multiply,diff,quotient, degree
```

```python
type(summ),type(multiply),type(diff),type(quotient), type(degree)
```

```python
float_c = float(c)
print(float_c, type(float_c))
```

```python
int_c = int(float_c)
print(int_c, type(int_c))
```

### Словари
Словари - это изменяемый упорядоченный тип данных:

данные в словаре - это пары ключ: значение

доступ к значениям осуществляется по ключу, а не по номеру, как в списках

данные в словаре упорядочены по порядку добавления элементов

так как словари изменяемы, то элементы словаря можно менять, добавлять, удалять

ключ должен быть объектом неизменяемого типа: число, строка, кортеж

значение может быть данными любого типа

```python
london_co = {
    'r1': {
        'hostname': 'london_r1',
        'location': '21 New Globe Walk',
        'vendor': 'Cisco',
        'model': '4451',
        'ios': '15.4',
        'ip': '10.255.0.1'
    },
    'r2': {
        'hostname': 'london_r2',
        'location': '21 New Globe Walk',
        'vendor': 'Cisco',
        'model': '4451',
        'ios': '15.4',
        'ip': '10.255.0.2'
    },
    'sw1': {
        'hostname': 'london_sw1',
        'location': '21 New Globe Walk',
        'vendor': 'Cisco',
        'model': '3850',
        'ios': '3.6.XE',
        'ip': '10.255.0.101'
    }
}
```

### Кортеж

Грубо говоря, кортеж - это список, который нельзя изменить. То есть, в кортеже есть только права на чтение. Это может быть защитой от случайных изменений. Константный список

```python
tuple1 = tuple()
tuple1
```

```python
tuple2 = ('ip',)
tuple2
```

```python
list_keys = ['hostname', 'location', 'vendor', 'model', 'ios', 'ip']
tuple_keys = tuple(list_keys)
tuple_keys
```

```python
tuple_keys[0]
```

МЫ не можем перезаписывать уже существующее значение в кортежах

```python
tuple_keys[0] = 'new_field'
```

Мы не можем добавлять через Append

```python
tuple_keys.append('new_field')
```

Функция sorted сортирует элементы кортежа по возрастанию и возвращает новый список с отсортированными элементами:



```python
sorted(tuple_keys)
```

### Множества 

Множество - это изменяемый неупорядоченный тип данных. В множестве всегда содержатся только уникальные элементы.

```python
numbers = [10, 20, 30, 40, 100, 10]
sett = set(numbers)
sett
```

```python
sett.add(90)
sett
```

```python
sett.discard(90)
sett
```

```python
10 in sett
```

```python
sett.clear()
sett
```

```python
10 in sett
```

### Булевы значения

Булевы значения в Python это две константы True и False.

В Python истинными и ложными значениями считаются не только True и False.

* истинное значение:
 * любое ненулевое число
 * любая непустая строка
 * любой непустой объект
* ложное значение:
 * 0
 * None
 * пустая строка
 * пустой объект
 
 
Остальные истинные и ложные значения, как правило, логически следуют из условия.

```python
bool(0),bool(1)
```
